```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

### Задание 1

```python
boston = load_boston()
boston.keys()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = boston['data']
data.shape
```

```python
feature_name = boston['feature_names']
feature_name
```

```python
target = boston['target']
target[:5]
```

```python
X = pd.DataFrame(data, columns=feature_name)
X.head()
```

```python
Y = pd.DataFrame(target, columns=['price'])
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_prediction = lr.predict(X_test)
Y_prediction.shape
```

```python
check_test = pd.DataFrame({
    "Y_test": Y_test["price"],
    "Y_pred": Y_prediction.flatten(),
})

check_test.head(10)
```

```python
check_test['error'] = check_test['Y_pred'] - check_test['Y_test']
check_test.head()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score_issure = r2_score(check_test["Y_pred"], check_test["Y_test"])
r2_score_issure
```

### Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
model.fit(X_train, Y_train.values[:, 0])
```

```python
y_pred = model.predict(X_test)
y_pred.shape
```

```python
check_test_1 = pd.DataFrame({
    "Y_test": Y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test_1.head(10)
```

```python
check_test_1['error'] = check_test_1['y_pred'] - check_test_1['Y_test']
check_test_1
```

```python
r2_score_issure_1 = r2_score(check_test_1['y_pred'], check_test_1['Y_test'])
r2_score_issure_1
```

Так как в первом задании мы получили в r2_score 0.669370269149561, а в r2_score_issure_1 0.8479049999699443
это значит (как я понимаю), данное решение подходит больше и показывает куда больше результат
